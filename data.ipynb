```python
import pandas as pd
import glob

files = glob.glob("/home/rayan/Downloads/archive (1)/*.csv")

dfs = []
for f in files:
    if "clean_world_happiness" in f:
        continue  
    
    df = pd.read_csv(f)
    
    year = f.split('/')[-1].split('.')[0]
    df['Year'] = int(year)
    
    df.columns = df.columns.str.strip().str.lower()
    
    rename_map = {
        'country': 'country',
        'region': 'region',
        'happiness rank': 'happiness_rank',
        'happiness.score': 'happiness_score',
        'happiness score': 'happiness_score',
        'score': 'happiness_score',
        'economy (gdp per capita)': 'gdp_per_capita',
        'economy..gdp.per.capita.': 'gdp_per_capita',
        'gdp per capita': 'gdp_per_capita',
        'family': 'social_support',
        'health (life expectancy)': 'life_expectancy',
        'health..life.expectancy.': 'life_expectancy',
        'healthy life expectancy': 'life_expectancy',
        'freedom': 'freedom',
        'freedom to make life choices': 'freedom',
        'trust (government corruption)': 'trust_government',
        'perceptions of corruption': 'trust_government',
        'generosity': 'generosity',
        'dystopia residual': 'dystopia_residual'
    }
    df.rename(columns=rename_map, inplace=True)
    
    keep_cols = [
        'country', 'region', 'year', 'happiness_rank', 'happiness_score',
        'gdp_per_capita', 'social_support', 'life_expectancy',
        'freedom', 'trust_government', 'generosity'
    ]
    
    for col in keep_cols:
        if col not in df.columns:
            df[col] = None

    df = df[keep_cols]
    dfs.append(df)

data = pd.concat(dfs, ignore_index=True)
data.fillna(data.mean(numeric_only=True), inplace=True)
data.drop_duplicates(inplace=True)

output_path = "/home/rayan/Downloads/archive (1)/clean_world_happiness.csv"
data.to_csv(output_path, index=False)

data.head()
```
